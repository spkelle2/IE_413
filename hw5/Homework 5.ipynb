```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Problem 1

```python
file_guy = 'parking.csv'
df = pd.read_csv(file_guy)
```

```python
df
```

The rate at which cars enter < the rate at which cars leave. Therefore, we immediately enter steady state and experience zero balks

### Part a

```python
balks = df['BALK'].values[:-1].max()
cars = df['CARS'].values[:-1].max()

balks/cars
```

### Part b

```python
# the number of parking spaces in use at time of an event
X = df['X'].values[:-1]

# maximum number of parking spaces ever in use at once
max_guy = int(X.max())

Y = range(max_guy+1)
```

```python
# total time exactly x parking spots are in use
time = np.zeros(max_guy+1)

for i in Y:
    # grab the last occurrance of x used spaces from X
    last_row = np.where(X==i)[0][-1]
    
    # grab the amount of time x spaces were used
    time[i] = df.iat[last_row,7]

time
```

```python
# the proportion of time exactly x spaces were used
props = time/time.sum()
props
```

```python
plt.style.use('ggplot')

plt.bar(Y, props, align='center')
plt.xlabel('Parking Spaces Filled')
plt.ylabel('Proportion of Time')
plt.suptitle('Distribution of Utilized Parking Spaces')
plt.show()
```

### Part c

```python
# expected number of spaces to be used at once
# proportion of time x spaces in use * x spaces = expected spaces used
expectation = (props*Y).sum()
expectation
```
